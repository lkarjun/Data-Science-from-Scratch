# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
years = [
    1950, 1960, 1970, 1980, 1990, 2000, 2010
]

gdp = [
    300.2, 543.3, 1075.9, 2862.5, 5979.6, 102089.7, 14958.3
]
```

```python
plt.plot(
    years, gdp,
    color = 'green', 
    marker = 'o',
    linestyle = 'solid'
)
plt.title("NOMINAL GDP")
plt.ylabel("BILLIONS OF $")
plt.show()
```

```python
# personal data
hour_spend = [
    6,4,5,5,6,4.2,5
]

days =[
    'Mon', 'Tue', 'wed', 'Thur', 'Fri', 'Sat', 'Sun'
]
```

```python
plt.plot(
    days, hour_spend,
    color = 'Red',
    marker = 'o',
    linestyle = 'solid'
)
plt.title('Hours Spending')
plt.ylabel('Hours')
plt.xlabel('days')
plt.show()
```

```python
value1 = [1,2,3,4,5]
value2 = ['mon', 'tues', 'wed', 'thurs']
plt.plot(
    value1,
    value2
)
```

# Bar Chart

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
plt.bar(range(len(movies)), num_oscars)
plt.title("My Favorite Movies")
plt.ylabel("# of Academy Awards")
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
plt.bar(range(len(days)), hour_spend)
plt.title("Personal data")
plt.ylabel("hours spend")
plt.xticks(range(len(days)), days)
plt.show()
```

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
print(histogram)
```

```python
print(histogram.values())
print(histogram.keys())
```

```python
plt.bar(
    [x+5 for x in histogram.keys()],
    histogram.values(),
    5,
    edgecolor = (1,1,1)
)

plt.axis([-5, 120, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('distribution of exam 1 grades')
plt.show()
```

```python
mentions = [500, 505]
years = [2017, 2018]
plt.bar(years, mentions, 0.5)
plt.xticks(years)
```

```python
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' increase")
plt.show()
```
