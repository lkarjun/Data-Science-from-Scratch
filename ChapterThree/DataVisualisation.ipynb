# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
years = [
    1950, 1960, 1970, 1980, 1990, 2000, 2010
]

gdp = [
    300.2, 543.3, 1075.9, 2862.5, 5979.6, 102089.7, 14958.3
]
```

```python
plt.plot(
    years, gdp,
    color = 'green', 
    marker = 'o',
    linestyle = 'solid'
)
plt.title("NOMINAL GDP")
plt.ylabel("BILLIONS OF $")
plt.show()
```

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
print(histogram)
```

```python
print(histogram.values())
print(histogram.keys())
```

```python
plt.bar(
    [x+5 for x in histogram.keys()],
    histogram.values(),
    5,
    edgecolor = (1,1,1)
)

plt.axis([-5, 120, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('distribution of exam 1 grades')
plt.show()
```

# Line Charts

```python
varience = [1,2,4,8,16,32,64,128,256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x, y in zip(varience, bias_squared)]
```

```python
xs = [i for i,_ in enumerate(varience)]
```

```python
plt.plot(
    xs,varience, 'g-',
    label = 'varience',
)
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')
plt.legend(loc=9)
plt.xlabel('Model complexity')
plt.title('The Bias- Variance Tradeoff')
plt.show()
```

# Scatterplots

```python
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                xy=(friend_count, minute_count),
                xytext=(5,-5),
                textcoords='offset points'
                )
```

# Alert! Axes aren't Comparable

```python
test1 = [99, 90, 73, 43, 64]
test2 = [100, 30, 40, 35, 37]
plt.scatter(test1, test2)
plt.title('Axes arent comparable')
plt.xlabel('test 1')
plt.ylabel('test 2')
```

```python
mark_of_lal = [40, 23, 65, 35, 99]
mark_of_arjun = [33, 66, 77, 88, 99]
sub = ['english', 'maths', 'science', 'computer', 'ai']
subject = [value.upper() for value in sub]
ss = [x for x,_ in enumerate(mark_of_lal)]

plt.plot(ss, mark_of_lal, 'g-', label='lal', marker='o', color='black')
plt.plot(ss, mark_of_arjun, 'b:', label='arjun', marker='o', color='red')
plt.legend(loc=9)
plt.title('Mark of students')
plt.ylabel('Marks', fontsize=15, color='blue')
plt.xticks(range(len(subject)),subject)
plt.tick_params(color='red',labelcolor='green',labelsize='large')
plt.show()
```

```python
lals_mark = [23,57,44,22,67]
arjuns_mark = [33,56,77,23,32]
plt.bar(
    range(len(lals_mark)),
    lals_mark,
    0.7,
    label='mark of lal',
    color='red'
)
plt.legend(loc=9)
plt.ylabel('Mark', color='green', fontsize=15)
plt.xticks(range(len(lals_mark)), lals_mark)
plt.title('display mark')
plt.show()
```
